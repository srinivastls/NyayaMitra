```python
from pymilvus import utility

# List all collections
collections = utility.list_collections()
print("Available Collections:")
for collection in collections:
    print(collection)
```

```python
from pymilvus import Collection

# Specify the collection name
collection_name = "JudgementCollection"  # Replace with your collection name

# Load the collection
collection = Collection(name=collection_name)

# Print the schema
print(f"Schema of collection '{collection_name}':")
print(collection.schema)
```

```python
# Load the collection to access its data
collection.load()

field_names = [field.name for field in collection.schema.fields]

# Retrieve a sample of data
results = collection.query(
    expr="",  # No filtering condition
    output_fields=field_names,  # List of field names to include in the query
    limit=30  # Adjust the limit to retrieve more data
)

print("Sample Data from Collection:")
for result in results:
    print(result)
```

```python
# Count the number of entities in the collection
entity_count = collection.num_entities
print(f"Total number of entities in collection '{collection_name}': {entity_count}")
```
