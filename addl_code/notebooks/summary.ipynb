```python
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
```

```python
from langchain_community.llms import Ollama

# Initialize the model
model_name = "phi3"
llm = Ollama(model=model_name)
```

```python
template = """Given the main section from an Indian legal document, 
generate a comprehensive list of all its subsections. The output should maintain the hierarchical structure, 
with each subsection clearly labeled and indented according to its level within the section. 
Ensure that the formatting is clean and organized, suitable for legal documentation. 
The subsections should reflect the logical flow of the main section, providing clarity and ease of reference. 
{context}
MAIN SECTION: {question}
Helpful Answer:"""
rag_prompt = PromptTemplate.from_template(template)
retriever =vector_store.as_retriever(search_type="mmr", search_kwargs={"k": 1})
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | rag_prompt
    | llm
)

print(rag_chain.invoke("Section 437 of H MC Act Award costs"))
```

```python

from langchain.schema.runnable import RunnablePassthrough
from functools import lru_cache

# Define the template
template = """Given the main section from an Indian legal document, 
generate a comprehensive list of all its subsections. The output should maintain the hierarchical structure, 
with each subsection clearly labeled and indented according to its level within the section. 
Ensure that the formatting is clean and organized, suitable for legal documentation. 
The subsections should reflect the logical flow of the main section, providing clarity and ease of reference. 
{context}
MAIN SECTION: {question}
Helpful Answer:"""
rag_prompt = PromptTemplate.from_template(template)

# Assume vector_store is already defined and connected to Milvus
retriever = vector_store.as_retriever(search_type="mmr", search_kwargs={"k": 1})

# Define the RAG chain
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | rag_prompt
    | llm
)

# Optionally cache results of the retrieval
@lru_cache(maxsize=100)
def cached_retriever(section):
    return rag_chain.invoke(section)

# Function to process a single section
def process_section(section):
    return cached_retriever(section)

# Example usage: processing a single section
result = process_section("Section 437 of H MC Act Award costs")
print(result)

# # To process multiple sections in parallel:
# sections = ["Section 437 of H MC Act Award costs", "Section 438 of XYZ Act Another Topic", "..."]

# # Option 1: Parallel processing
# with ThreadPoolExecutor() as executor:
#     results = list(executor.map(process_section, sections))

# # Option 2: Sequential processing (if parallelization is not needed)
# # results = [process_section(section) for section in sections]

# # Print or process the results
# for res in results:
#     print(res)
```
