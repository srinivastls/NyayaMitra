```python
import fitz  # PyMuPDF
from transformers import pipeline
import pymupdf

def extract_text_from_pdf(file_path):
    # Open the provided PDF file
    document =  pymupdf.open(file_path)
    text = ""
    
    # Extract text from each page
    for page in document:
        text += page.get_text()
    
    document.close()
    return text

def process_text_with_llamaparse(text):
    # Load LlamaParse model
    llamaparse = pipeline("structured-data-extraction", model="Salesforce/llamaparse-large")

    # Process the extracted text
    results = llamaparse(text)
    return results

def save_extracted_data(data, output_file):
    # Save the extracted data into a new file
    with open(output_file, 'w') as f:
        for item in data:
            f.write(f"Heading: {item['label']}\n")
            f.write(f"Content: {item['text']}\n\n")


```

```python
# Specify the file paths
pdf_file_path = "/home/ubuntu/ram/code/data/Greater_Hyderabad_Municipal_Corporation_Act_1955.PDF"
output_file_path = "extracted_data.txt"

# Extract text from PDF
extracted_text = extract_text_from_pdf(pdf_file_path)

# Process text to extract structured data
structured_data = process_text_with_llamaparse(extracted_text)

# Save the structured data
save_extracted_data(structured_data, output_file_path)
```
