## supreme court extraction

```python
import re
import pdfplumber
import pandas as pd

end_heading = 'JUDGMENT:'

def parse_pdf_pdfplumber(file_path):
    #text = ""
    # Open the PDF
    try:
        # Open the PDF file
        doc=pdfplumber.open(file_path)
    except FileNotFoundError:
        print(f"The file {pdf_path} was not found.")
        return None,None
    except Exception as e:
        print(f"An error occurred while opening the file: {e}")
        return None,None
    
    text = "THE SUPREME COURT OF INDIA \n"
    text2=""
    extracting = True
    extract2=False
    extract=True
    flag=0
    for page in doc.pages:
        page_text = page.extract_text()
        lines = page_text.split('\n')
        for line in lines:
            if (not re.match(r'^\d|http', line)) and line!="SUPREME COURT OF INDIA" and extract2 and (not re.match(r'Page', line)) and len(line)!=0 :
                flag=1
                text2 +=line +'\n'
            if end_heading ==line :
                extracting = False
                extract2=True
                 # Stop extracting after the end heading
            if(line=="CITATION:" or line== "ACT:"):
                extracting=False
            if (not re.match(r'^\d|http', line)) and line!="SUPREME COURT OF INDIA" and extracting and (not re.match(r'Page', line)) and len(line)!=0 and not extract2:
               
               text += line + '\n'
    if(flag==0):
        return None,None
            
            
    return text.strip(),text2.strip()


text_pdfplumbers=[]
text_judgments=[]
for i in files[31468:]:
    print(i)
    pdf_path='supreme/'+i
    text_pdfplumberm,text_judg = parse_pdf_pdfplumber(pdf_path)
    if(text_pdfplumberm !=None and text_judg!=None):
        text_pdfplumbers.append(text_pdfplumberm)
        #print(text_pdfplumberm)
        text_judgments.append(text_judg)

df=pd.DataFrame()
df['meta data']=text_pdfplumbers
df['judgments']=text_judgments
df.to_csv("supreme2.csv",index=False)
```
